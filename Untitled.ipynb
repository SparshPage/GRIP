```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris_data = pd.read_csv('Iris.csv')
iris_data.head()
X = iris_data[["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]]
X
```

```python
y_un = iris_data['Species'].unique()
iris_data['Species'] = iris_data.Species.apply(lambda x : list(y_un).index(x))
iris_data
```

```python
#Some EDA on the data

import seaborn as sns
sns.jointplot(data = iris_data , x = 'SepalLengthCm', y = 'SepalWidthCm')
```

```python
fig, Ax = plt.subplots(2,1, figsize = (10, 10), sharey = False)

sns.kdeplot(ax = Ax[0]  ,data = iris_data,  x = 'SepalLengthCm', y = 'SepalWidthCm')
sns.kdeplot(ax = Ax[1],data = iris_data,  x = 'PetalLengthCm', y = 'PetalWidthCm')


```

```python
#Lets find out which features affect the classification the most
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
y =iris_data['Species']


Xtrain, Xtest, ytrain, ytest = train_test_split(X,y, test_size = 0.2)

clf  =RandomForestClassifier(n_estimators = 1000).fit(Xtrain,ytrain)
y_pred = clf.predict(Xtest)
accuracy_score(ytest, y_pred)
```

```python
feature_imp = pd.Series(clf.feature_importances_, iris_data.columns[[1,2,3,4]]).sort_values(ascending = False)
sns.barplot(x = feature_imp, y = feature_imp.index)
```

```python
from sklearn.cluster import KMeans
iris_data['Species'].unique()
X = iris_data.iloc[:, [1,2,3,4]].values

preds = []

for i in range(1,11):
    kmeans = KMeans(n_clusters = i, random_state = 0).fit(X)
    
    preds.append(kmeans.inertia_)
    
    
  
   
plt.plot(range(1,11), preds)    
```

```python
fig = plt.figure("Kmeans", figsize = (10,10))
ax = fig.add_subplot(111, projection = '3d', )

km = KMeans(n_clusters = 3, init = 'k-means++').fit(X)
y_kmeans= km.fit_predict(X)

ax.scatter(X[y_kmeans == 0, 2], X[y_kmeans == 0, 3], X[y_kmeans == 0,1],
            s = 100, c = 'red', label = 'Iris-setosa')
ax.scatter(X[y_kmeans == 1, 2], X[y_kmeans == 1, 3], X[y_kmeans == 1,1],
            s = 100, c = 'blue', label = 'Iris-versicolour')
ax.scatter(X[y_kmeans == 2, 2], X[y_kmeans == 2, 3],X[y_kmeans == 2,1],
            s = 100, c = 'green', label = 'Iris-virginica')


ax.set_xlabel("Petal Length")
ax.set_ylabel("Petal Width")
ax.set_zlabel("Sepal Length")

#plt.scatter(km.cluster_centers_[:,0], km.cluster_centers_[:,1],s = 100, c = 'yellow', label = 'Center')
plt.legend()
```
